```python
from matplotlib import pyplot as plt
```

```python
filename = "./DenseMap/DenseMap_0250_01"
output = "./DenseMapReadyToRun/DenseMap_0250_01.txt"
f = open(output,'w')
x = []
y = []
i = 1
for line in open(filename):
    if line[0].isdigit():
        # print(line)
        ss = line.split(" ")
        x.append(float(ss[0]))
        y.append(float(ss[1]))
        res = str(i)+" "+str(ss[0])+" "+str(ss[1])+"\n"
        print(res)
        i = i + 1
        f.write(res)
print(x[:5], y[:5])
```

```python
fig1 = plt.figure(figsize=(16,9))
plt.scatter(x,y,s=2)
print(len(x))
print(min(x),max(x))
print(min(y),max(y))
```
