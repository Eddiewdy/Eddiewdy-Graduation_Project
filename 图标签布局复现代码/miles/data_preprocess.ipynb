```python
output = "out.txt"
file = open('./readytorun_withname.txt','w')
index = 1
for line in open(output):
    l = line.index('[')
    r = line.index(']')
    n = line.index(',')
    trim = line[l+1:r]
    # print(trim)
    res = line[:n]+","+trim+"\n"
    print(res)
    file.writelines(res)
    index = index + 1
```

```python
import numpy as np
point=[]
for line in open('./readytorun.txt'):
    index, x, y = line.strip().split(",")
    point.append([float(x),float(y)])
print(point)
arr = np.array(point)
```
